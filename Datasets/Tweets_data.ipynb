## Cleaning data and extracting equal rows

```python
import pandas as pd
from langdetect import detect, LangDetectException
import re
import unicodedata
```

```python
def clean_text(text):
    # Remove words starting with '@'
    cleaned_text = re.sub(r'@\w+\s?', '', text)
    
    # Remove URLs
    cleaned_text = re.sub(r'http\S+|www\S+', '', cleaned_text)
    
    # Remove HTML entities
    cleaned_text = re.sub(r"&\w+;", '', cleaned_text)
    
    # Remove Unicode characters and normalize the text
    cleaned_text = unicodedata.normalize('NFKD', cleaned_text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
    
    # Remove special characters
    cleaned_text = re.sub(r"[^a-zA-Z0-9\s]", '', cleaned_text)
    
    # Remove extra whitespaces
    cleaned_text = re.sub(r"\s+", ' ', cleaned_text).strip()
    
    return cleaned_text
```
